```
import cv2
from google.colab.patches import cv2_imshow
```

```
image1=cv2.imread('scenary-1.jpg')
image2=cv2.imread('scenary-2.jpg')
image3=cv2.imread('scenary-3.jpg')
```

```
#resizing image2 to become equal to image1
image2 = cv2.resize(image2,(300,168), interpolation = cv2.INTER_AREA) 
```

```
#Now cropping image1 and image2
crop_img1=image1[0:,0:150]
crop_img2=image2[0:,0:150]
```

```
cv2_imshow(crop_img1)
cv2_imshow(crop_img2)
```

```
#to concatenate two images
import numpy as np
images=np.concatenate((crop_img1,crop_img2),axis=1)
```

```
cv2_imshow(images)
```
